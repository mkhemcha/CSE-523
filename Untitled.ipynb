```python
from collections import Counter
from nltk.corpus import stopwords
import re
import string
import pandas as pd
import math
from nltk.stem.lancaster import LancasterStemmer


emoticons_str = r"""
    (?:
        [:=;] # Eyes
        [oO\-]? # Nose (optional)
        [D\)\]\(\]/\\OpP] # Mouth
    )"""
 
regex_str = [
    emoticons_str,
    r'<[^>]+>', # HTML tags
    r'(?:@[\w_]+)', # @-mentions
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)", # hash-tags
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&amp;+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+', # URLs
 
    r'(?:(?:\d+,?)+(?:\.?\d+)?)', # numbers
    r"(?:[a-z][a-z'\-_]+[a-z])", # words with - and '
    r'(?:[\w_]+)', # other words
    r'(?:\S)' # anything else
]
    
tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)
 
def tokenize(s):
    return tokens_re.findall(s)
 
def preprocess(s, lowercase=False):
    tokens = tokenize(s)
    if lowercase:
        tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
    return tokens

def tf_idf_weight(data,vocab,totaldocs):
    result=dict()
    count1=Counter(data)
    count2=Counter(list(vocab))
    for key,value in count1.items():
        t=(value/len(data))*(math.log10(totaldocs/count2[key]))
        result[key]=t
    return result

input_data=pd.read_table('train.tsv',sep= '\t',names=['Col1','PlaceId','Col2','Col3','Name','Col5','Col6','Col7','Col8','Col9','Col10','Col11','Col12','Col13','data'])
input_data.dropna(axis=0,how='all',subset=['data'],inplace=True)
# input_data.groupby('PlaceId')['data']
groupby=input_data.groupby('PlaceId')
# per place id(collection of tweet per place id one doc)
totaldocs=0
vocab=[]
features=dict()
lancaster_stemmer = LancasterStemmer()
for name,group in groupby:
    collect_data=[]
    hash_data=[]
    count=0
    for index,row in group.iterrows():
        tokens=preprocess(row['data'])
        if count==0:
            place=row['Name'].split(',')
            count+=1
            for item in place:
                if('name:' in item):
                    index=item.index('name:')
                    placename=item[index+6:]
                    break
            print(placename)
        terms_hash = [lancaster_stemmer.stem(term) for term in tokens if term.startswith('#')]
        punct=list(string.punctuation)
        stoplist = stopwords.words('english') + punct
        relevant_tokens=[lancaster_stemmer.stem(token) for token in tokens if token not in stoplist and not token.startswith(('#','@')) and token.isalnum()==True and len(token) >= 2]
        collect_data.append(relevant_tokens)
        hash_data.append(terms_hash)
    flat_hash= [item for sublist in hash_data for item in sublist]
    totaldocs+=1
    flat_list = [item for sublist in collect_data for item in sublist]
    for item in flat_hash:
        flat_list.append(item.replace("#",""))
    vocab.append(list(set(flat_list)))
    features[name]=flat_list
vocab=[item for sublist in vocab for item in sublist]
for key,value in features.items():
    result=tf_idf_weight(value,vocab,totaldocs)
    print("Place Id : " + str(key))
    print(dict(Counter(result).most_common(5)))
```
